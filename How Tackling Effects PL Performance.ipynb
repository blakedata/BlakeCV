```python
#Building on this to create a simple visualization

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")

# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(6, 15))
# Load data from a CSV file into a Pandas DataFrame
dataFrame = pd.read_csv('C:\\Users\\118358981\\Downloads\\PLStats22.23 - Sheet1.csv').sort_values("Tackles", ascending=False)

  
# creating a rank column and passing the returned rank series
# change method to 'min' to rank by minimum
dataFrame["Rank"] = dataFrame["Tackles"].rank(method ='min')

# Plot 
sns.set_color_codes("pastel")
sns.barplot(x="Rank", y="Squad", data=dataFrame,
            label="Tackle Rank", color="b")


ax.set(xlim=(0, 20), ylabel="Squad",
       xlabel="Tackle Ranking by Team", )
# ax.bar_label(ax.containers[0])
sns.despine(left=True, bottom=True)
```
